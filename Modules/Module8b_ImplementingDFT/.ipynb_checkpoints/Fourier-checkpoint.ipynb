# Discovering The Fourier Transform

```python
import numpy as np
import matplotlib.pyplot as plt
```

In this module, you're going to explore sums of products of sinusoids and make some observations.  This will form the "basis" of Fourier analysis, the most important analytical tool we'll use in this course.

The code below sets up two waveforms and and plots each one, as well as the product of the two.  It also reports the sum of all of the samples in the product plot.  At the moment, they are 2hz and 3hz, respectively, and the sum of their product is an incredibly small number (which you can consider to be numerically zero).  You will modify the code to try different signals.  In every example you look at, the signal will go through an integer number of periods.

Given this setup, examine the following combinations of $y_1(t)$ and $y_2(t)$, and report the sums:

1. $y_1(t)$ is a 2hz cosine and $y_2(t)$ is a 4hz cosine

2. $y_1(t)$ is a 2hz cosine and $y_2(t)$ is also a 2hz cosine

3. $y_1(t)$ is a 3hz cosine and $y_2(t)$ is also a 3hz cosine

4. $y_1(t)$ is a 3hz cosine and $y_2(t)$ is a 3hz sine


```python
n_samples = 100
t = np.linspace(0, 1, n_samples+1)[0:n_samples]
f1 = 2
f2 = 4
y1 = np.cos(2*np.pi*f1*t)
y2 = np.cos(2*np.pi*f2*t)

plt.figure(figsize=(12, 12))
prod = y1*y2
plt.subplot(311)
plt.stem(t, y1)
plt.ylim([-1.1, 1.1])
plt.title("$y_1(t)$")

plt.subplot(312)
plt.stem(t, y2)
plt.ylim([-1.1, 1.1])
plt.title("$y_2(t)$")

plt.subplot(313)
plt.stem(t, prod)
plt.ylim([-1.1, 1.1])
plt.title("$y_1(t)y_2(t)$")
plt.show()

print("Sum = ", np.sum(prod))
```
