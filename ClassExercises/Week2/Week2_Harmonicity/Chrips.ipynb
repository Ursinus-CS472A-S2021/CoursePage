```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
```

## $g(t) = \cos(2 \pi f(t)) $

## Ex) $f(t) = t + 10$
## $g(t) = \cos(2 \pi (t + 10)) = \cos(2 \pi t + 20 \pi) $
## Ex) $ f(t) = 440*t^2 $
## $ g(t) = \cos (2 \pi t^2) $

## Ex) $y(t) = \cos(2 \pi f * t) $
## $g(t) = f*t $

```python
sr = 44100
f = 440*(t**2)
t = np.arange(sr)/sr
y = np.cos(2*np.pi*f)
ipd.Audio(y, rate=sr)
```

```python
f2 = 440*(t**3)
t = np.arange(sr)/sr
y = np.cos(2*np.pi*f2)
ipd.Audio(y, rate=sr)
```

```python
plt.plot(t, f)
plt.plot(t, f2)
plt.legend(["$t^2$", "$t^3$"])
```
