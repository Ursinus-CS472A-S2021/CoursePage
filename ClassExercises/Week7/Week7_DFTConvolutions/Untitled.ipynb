```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
from scipy.signal import fftconvolve
import librosa

x, sr = librosa.load("robot.wav")
h = np.zeros(sr)
T = 1000
#h[0:T*5:T] = 1
h[0:T] = 1
plt.plot(h)
plt.xlim([0, T*21])

y = fftconvolve(x, h)
```

```python
X = np.fft.fft(x)
nfreqs = int(len(x)/2) # Only plot the first half of the
# frequencies since the second half is the mirror image_on
X = X[0:nfreqs]
freqs = np.arange(nfreqs)*sr/len(x)
plt.plot(freqs, np.abs(X))
ipd.Audio(x, rate=sr)
plt.xlim([0, 2000])
```

```python
H = np.fft.fft(h)
nfreqs = int(len(h)/2) # Only plot the first half of the
# frequencies since the second half is the mirror image_on
H = H[0:nfreqs]
freqs = np.arange(nfreqs)*sr/len(h)
plt.plot(freqs, np.abs(H))
plt.xlim([0, 2000])
```

```python
Y = np.fft.fft(y)
nfreqs = int(len(y)/2) # Only plot the first half of the
# frequencies since the second half is the mirror image_on
Y = Y[0:nfreqs]
freqs = np.arange(nfreqs)*sr/len(y)
plt.plot(freqs, np.abs(Y))
plt.xlim([0, 2000])
ipd.Audio(y, rate=sr)
```
